```python
# menghilangkan kolom data yang tidak diperlukan sebelum diproses lebih lanjut
df = df.drop(["#", "Name", "Generation"], axis=1)
df
```

```python
# mencari apakah terdapat data yang Null (missing values)
df.isna().sum()
```

```python
import numpy as np

# missing values dihilangkan dengan menyisipkan data dari kolom sebelumnya, yaitu kolom "Type 1"
df["Type 2"].fillna(df["Type 1"], inplace=True)
df
```

```python
# melakukan pengecekan kembali terhadap missing values
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# mengganti datatype object menjadi integer
le = LabelEncoder()

df["Type 1"] = le.fit_transform(df["Type 1"])
df["Type 2"] = le.fit_transform(df["Type 2"])
df["Legendary"] = le.fit_transform(df["Legendary"])

df
```

```python
# menampilkan korelasi data dari setiap kolom
corr = df.corr()
corr
```

```python
# menampilkan korelasi data dari kolom "Legendary" saja
corrWith = df.corrwith(df["Legendary"])
corrWith
```

```python
# menampilkan heatmap dari korelasi yang sudah ditampilkan
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f')
```

```python
# memisahkan data x dan y dari dataset
y = df["Legendary"]
x = df.drop("Legendary", axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

# melakukan standarisasi data x menggunakan StandardScaler
scaler = StandardScaler()

scaler.fit(x)

x = scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split

# melakukan splitting data dengan rasio data test sebesar 20% dari keseluruhan data
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# mencari nilai k yang paling sesuai untuk dilakukan klasifikasi menggunakan KNN
score = []

for i in range(2,15):
  knn = KNeighborsClassifier(i)
  knn.fit(xTrain, yTrain)
  score = knn.score(xTest, yTest)

  print(f"K: {i}, KScore: {score}")
```

```python
# membuat model klasifikasi data menggunakan KNN dengan nilai k terbesar yaitu k=2
knn = KNeighborsClassifier(n_neighbors=2)

knn.fit(xTrain, yTrain)

yPred = knn.predict(xTest)
```

```python
from sklearn.metrics import precision_score, recall_score, ConfusionMatrixDisplay, classification_report, accuracy_score, f1_score

# menampilkan confusion matrix dari hasil prediksi terhadap data test, yaitu label asli dan label prediksi
ConfusionMatrixDisplay.from_predictions(yTest, yPred, display_labels=["False", "True"])

plt.show()
```

```python
# menampilkan skor accuracy, precision, recall, dan f1 dari model
print(f"Accuracy Score: {accuracy_score(yTest, yPred)}")
print(f"Precision Score: {precision_score(yTest, yPred)}")
print(f"Recall Score: {recall_score(yTest, yPred)}")
print(f"F1 Score: {f1_score(yTest, yPred)}")
```

```python
# menampilkan skor menggunakan function classification_report
print(classification_report(yTest, yPred))
```
